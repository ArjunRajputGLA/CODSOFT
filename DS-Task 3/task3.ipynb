IRIS FLOWER CLASSIFICATION:

The Iris flower dataset consists of three species: setosa, versicolor, and virginica. These species can be distinguished based on their measurements. Now, imagine that you have the measurements of Iris flowers categorized by their respective species. Your objective is to train a machine learning model that can learn from these measurements and accurately classify the Iris flowers into their respective species.

Use the Iris dataset to develop a model that can classify iris flowers into different species based on their sepal and petal measurements. This dataset is widely used for introductory classification tasks.


```python
import pandas as pd
df = pd.read_csv('IRIS.csv')
df.head()
```

```python
display(df.info())
display(df.shape)
```

```python
display(df.isnull())
display(df.isnull().sum())
```

```python
display(df['species'].unique())
display(df['species'].nunique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(12, 10), dpi=200)

sns.histplot(data=df, x='sepal_length', hue='species', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(data=df, x='sepal_width', hue='species', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(data=df, x='petal_length', hue='species', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(data=df, x='petal_width', hue='species', kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout()
plt.show() 
```
